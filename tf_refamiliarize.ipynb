```python
import tensorflow as tf
import numpy as np
import random
```

```python
train_data = np.arange(30, dtype=int)  # x + 36
train_label = np.arange(30, dtype=int) + 36
shuffled_data = list(zip(train_data, train_label))
random.shuffle(shuffled_data)
shuffled_data = [[i for i, j in shuffled_data], [j for i, j in shuffled_data]]
train_data = np.asarray(shuffled_data[0], dtype=int)
train_label = np.asarray(shuffled_data[-1], dtype=int)

model = tf.keras.Sequential([
    # tf.keras.layers.Dense(units=8, input_shape=(1,)),
    tf.keras.layers.Input(1),
    tf.keras.layers.Dense(units=64, input_shape=(1,)),
    tf.keras.layers.Dense(units=64),
    tf.keras.layers.Dense(units=1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-1), loss=tf.keras.losses.mse)

model.fit(train_data, train_label, epochs=500)
```

```python
print(np.round(model.predict([9001])[0][0]))
```
